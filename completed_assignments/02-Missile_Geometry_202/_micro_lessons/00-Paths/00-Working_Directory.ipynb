
## 00 Working Directory

## Goal

- Every program runs from a specific folder called the **working directory**.
- Before loading data files, we must know where Python is running.

## Why this matters
  - When Python tries to open a file, it begins from the working directory unless told otherwise.
  - If you do not understand this idea, every future `FileNotFoundError` will become a dramatic stage production.

````python
from pathlib import Path

cwd = Path.cwd()

print("Current Working Directory:")
print(cwd)
````

## Concept

- Think of the working directory as the **starting location** for relative paths.

- **If Python sees:**
  - Path("data/countries.geojson")

- **It interprets that as:**
  - Path.cwd() / "data/countries.geojson"


````python
print("Items in the current working directory:")
for item in Path.cwd().iterdir():
    print(" -", item.name)
````

## Exercise B

List the contents of the *parent* directory — one level above where Python is running.

Use `.parent` on `Path.cwd()` and call `.iterdir()` on the result.

````python
from pathlib import Path

for item in Path.cwd().parent.iterdir():
    print(item)
````

## Optional Advanced — Changing the Working Directory

`os.chdir()` can change the working directory mid-notebook.

1. Change to the parent directory using `os.chdir(Path.cwd().parent)`
2. Print the new `Path.cwd()`
3. Change back to the original

Then answer: what risks does calling `os.chdir()` mid-notebook introduce for other cells?

Changing the working directory can break other cells that rely on relative paths.

````python
import os
from pathlib import Path

original = Path.cwd()

# 1. Change to the parent directory
os.chdir(Path.cwd().parent)

# 2. Print the new cwd
print(Path.cwd())

# 3. Change back to original
os.chdir(original)

print(Path.cwd())
````
