# 📘 01 - Scalar Types

Review and practice working with Python's basic scalar data types.

## ✅ Learning Goals
- Recognize scalar types (`int`, `float`, `str`, `bool`)
- Use the `type()` function to identify types
- Use variables and assignment

````python
# 🔧 Define some variables
num = 42
pi = 3.14159
name = 'Python'
is_cool = True
````

````python
# 🔍 Check their types
print(type(num))
print(type(pi))
print(type(name))
print(type(is_cool))
````

### ✏️ Your Turn
Create your own variables of each type and print them.

````python
age = 21
height = 5.1
name = "Andrea"
is_student = True

print(type(age), (age))
print(type(height), (height))
print(type(name), (name))
print(type(is_student), (is_student))   
````

## ⚖️ Comparison Practice
Use comparison operators to check values.

````python
x = 10
y = 5
print(x > y)
print(x == y)
print(x != y)
````
